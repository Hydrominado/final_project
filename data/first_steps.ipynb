```python
cuisine_reviews = df.groupby('cuisines')[['very_good', 'excellent', 'average', 'poor', 'terrible']].sum()
```

```python
cuisine_reviews_country = df.groupby(['country', 'cuisines'])[['very_good', 'excellent', 'average', 'poor', 'terrible']].sum()
```

```python
df.replace(['England','Wales','Scotland','Northern Ireland'],'United Kingdom', inplace=True)
```

```python
df.replace('Czech Republic','Czechia', inplace=True)
```

```python
df.replace('The Netherlands','Netherlands', inplace=True)
```

```python
total_reviews = df.groupby(['country', 'cuisines'])[['very_good', 'excellent', 'average', 'poor', 'terrible']].sum()
```

```python
total_reviews
```

```python
percentage_reviews = total_reviews.div(total_reviews.sum(axis=1), axis=0) * 100
```

```python
percentage_reviews['score']=(percentage_reviews['very_good']*5+percentage_reviews['excellent']*4+percentage_reviews['average']*3+percentage_reviews['poor']*2+percentage_reviews['terrible']*1)/100
```

```python
percentage_reviews.reset_index(inplace=True)
```

```python
percentage_reviews
```

```python
list(world.index)
```

```python
TypeFrench = percentage_reviews[percentage_reviews.cuisines=='French']
```

```python
TypeFrench
```

```python
TypeFrench.shape
```

```python
TypeFrench.set_index('country', inplace=True)
```

```python
cuisine_french=world.join(TypeFrench, how='inner')
```

```python
cuisine_french
```

```python
!pip3 install matplotlib
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
cuisine_french.plot(column='score', ax=ax, legend=True,
            legend_kwds={'label': "Score sur 5"},
            cmap='Blues')
ax.set_title('French')
ax.set_xlim(-20,40)
ax.set_ylim(30,80)
plt.show()
```
